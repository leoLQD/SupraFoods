# Cria Arquivo de Produtos

```python
def dataAno(xCod):
    return xCod[0:4]
```

```python
def dataMes(xCod):
    return xCod[5:7]
```

```python
# Classifica a Familia de Produtos em Função do CÓDIGO
def famProdutos(xCod):
    x = str(xCod)
    yCod = x[:2]
    if yCod == '01':
        return 'Blister'
    elif yCod =='02':
        return 'Temperos'
    elif yCod =='03':
        return 'Molhos'
    elif yCod == '04':
        return 'Conservas'
    elif yCod =='05': 
        return 'Institucional'
    elif yCod == '06':
        return 'Diversos '
    elif yCod =='07':
        return 'Insumos'
    elif yCod =='08':
        return 'Consumo'
    elif yCod == '09':
        return 'Produtos Prontos para Venda'
    elif yCod =='10':
        return 'Serviço'
    elif yCod == '11':
        return 'Embalagens'
    elif yCod =='12':
        return 'Contas Fixas'
    elif yCod == '13':
        return 'Bem Ativo'
    elif yCod =='14':
        return 'Manutenção de Ativo'
    elif yCod =='15':
        return 'Display'
    elif yCod == '16':
        return 'Supra Life'

```

###### Carrega Tabela de Produtos

```python
nomes = pd.read_csv('supraProdutos.csv', dtype={'cod': object, 'descricao': object}, encoding='latin-1')
```

```python
# Acrescenta nome da Familia de Produtos
nomes['Familia'] = nomes['cod'].apply(famProdutos)
```
