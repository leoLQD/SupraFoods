```python
from nfelib.v3_10 import leiauteNFe as leiauteNFe3
from datetime import datetime
import xlsxwriter
import os
from lxml import etree
import tempfile
os.chdir('C:\\\\Mettricx\\Projetos\\\\SupraAlimentos\\\\bkp_xml')
```

```python

# Create a workbook and add a worksheet.
workbook = xlsxwriter.Workbook('Vendas.xlsx')
worksheet = workbook.add_worksheet('Produtos')


# Add a bold format to use to highlight cells.
bold = workbook.add_format({'bold': 1})

# Add a number format for cells with money.
money_format = workbook.add_format({'num_format': '$#,##0'})

# Add an Excel date format.
date_format = workbook.add_format({'num_format': 'mmmm d yyyy'})

# Adjust the column width.
#worksheet.set_column(10, 10, 15, 30, 20, 5, 40, 5, 2, 10)
# Write some data headers.
worksheet.write('A1', 'NF', bold)
worksheet.write('B1', 'Data', bold)
worksheet.write('C1', 'CNPJ', bold)
worksheet.write('D1', 'Cidade', bold)
worksheet.write('E1', 'Bairro', bold)
worksheet.write('F1', 'UF', bold)
worksheet.write('G1', 'UF', bold)
worksheet.write('H1', 'Descrição', bold)
worksheet.write('I1', 'Qty', bold)
worksheet.write('J1', 'Unidade', bold)
worksheet.write('K1', 'Preço', bold)
```
